## 4. Comparing different Machine Learning Models

```python
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
models = {
      'KNN': KNeighborsClassifier(),
      'Logistic R': LogisticRegression(),
      'Random Forest': RandomForestClassifier()
  }
def fit_and_score(models, X_train, X_test, y_train, y_test):
  np.random.seed(42)
  model_score = {}
  for name, model in models.items():
    model.fit(X_train, y_train)
    model_score[name] = model.score(X_test, y_test)
  return model_score
```

```python
from sklearn.model_selection import cross_val_score

model_score = fit_and_score(
    models = models,
    X_train = X_train,
    X_test = X_test,
    y_train = y_train,
    y_test = y_test
)
model_score
```

```python
model_compare = pd.DataFrame(model_score, index = ['accuracy'])
model_compare
```

```python
model_compare.T.plot.bar();
plt.xticks(rotation = 0)
```

```python
train_score = []
test_score = []

neighbors = range(1,21)

knn = KNeighborsClassifier()

for i in neighbors:
  knn.set_params(n_neighbors = i)
  knn.fit(X_train, y_train)
  train_score.append(knn.score(X_train, y_train))
  test_score.append(knn.score(X_test, y_test))

```

```python
train_score
```

```python
test_score
```

```python
plt.plot(neighbors, train_score, label = 'Train Score')
plt.plot(neighbors, test_score, label = 'Test Score')

plt.xticks(np.arange(1,21,1))
plt.xlabel('Number of neighbours')
plt.ylabel('Model Score')
plt.legend()
print(f"Maximum KNN score on the test data: {max(test_score)*100:.2f}%")
```

## 5. Hyperparameter Tuning


 5.1. Tuning models with with RandomizedSearchCV

RandomizedSearchCV for LogisticRegression Model

```python
logis_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}

rf_grid = {"n_estimators": np.arange(10, 1000, 50),
           "max_depth": [None, 3, 5, 10],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2)}
```

```python
from sklearn.model_selection import RandomizedSearchCV

np.random.seed(42)
logis_grid = RandomizedSearchCV(LogisticRegression(), param_distributions= logis_grid, cv =5, n_iter=20, verbose =True)
logis_grid.fit(X_train, y_train);
```

5.2. Tuning Models with GridSearchCV

The difference between RandomizedSearchCV and GridSearchCV is:

sklearn.model_selection.RandomizedSearchCV searches over a grid of hyperparameters performing n_iter combinations (e.g. will explore random combinations of the hyperparameters for a defined number of iterations).
sklearn.model_selection.GridSearchCV will test every single possible combination of hyperparameters in the grid (this is a thorough test but can take quite a long time).


```python
from sklearn.model_selection import GridSearchCV
# Different LogisticRegression hyperparameters
gs_logis_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}
logistic_grid = GridSearchCV(LogisticRegression(), param_grid=gs_logis_grid, cv=5, verbose=True)
logistic_grid.fit(X_train, y_train);
```

```python
logistic_grid.best_params_
```

```python
logistic_grid.score(X_test, y_test)
```

6.1 ROC Curve and AUC Scores:


It's a way of understanding how your model is performing by comparing the true positive rate to the false positive rate.

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(estimator = logistic_grid, X=X_test, y=y_test);
```

This is great, our model does far better than guessing which would be a line going from the bottom left corner to the top right corner, AUC = 0.5.

But a perfect model would achieve an AUC score of 1.0, so there's still room for improvement.

6.2 Creating a confusion matrix

A confusion matrix is a visual way to show where your model made the right predictions and where it made the wrong predictions (or in other words, got confused).

```python
print(confusion_matrix(y_test, y_preds))
```

```python
# Import Seaborn
import seaborn as sns
sns.set(font_scale=1.5) # Increase font size

def plot_conf_mat(y_test, y_preds):
    fig, ax = plt.subplots(figsize=(3, 3))
    ax = sns.heatmap(confusion_matrix(y_test, y_preds),
                     annot=True, # Annotate the boxes
                     cbar=False)
    plt.xlabel("true label")
    plt.ylabel("predicted label")

plot_conf_mat(y_test, y_preds)
```

6.3 Classification report

A classification report is a collection of different metrics and other details.

```python
print(classification_report(y_test, y_preds))
```

SO this above classificaion_report is generated on 20% Test data. So now we need to do Cross Validation to get better scores.

6.4 Cross-Validation

```python
logistic_grid.best_params_
```

```python
from sklearn.model_selection import cross_val_score

# Instantiate best model with best hyperparameters (found with GridSearchCV)
clf = LogisticRegression(C=0.23357214690901212, solver = 'liblinear')

#Now we've got an instantiated classifier, let's find some cross-validated metrics.
```

```python
%%time
cv_acc = cross_val_score(clf, X, y, cv=5, scoring = "accuracy")
cv_acc
```

```python
#Wow we are getting Good score in Cross Val
#Since there are 5 outputs, we will take average of them

cv_acc = np.mean(cv_acc)
cv_acc
```

Now we'll do the same for other classification metrics.

```python
# Cross-validated precision score
cv_precision = np.mean(cross_val_score(clf,
                                       X,
                                       y,
                                       cv=5, # 5-fold cross-validation
                                       scoring="precision")) # precision as scoring
cv_precision
```

```python
# Cross-validated recall score
cv_recall = np.mean(cross_val_score(clf,
                                    X,
                                    y,
                                    cv=5, # 5-fold cross-validation
                                    scoring="recall")) # recall as scoring
cv_recall
```

```python
# Cross-validated F1 score
cv_f1 = np.mean(cross_val_score(clf,
                                X,
                                y,
                                cv=5, # 5-fold cross-validation
                                scoring="f1")) # f1 as scoring
cv_f1
```

#Visualizing Cross Validation Metrics

```python
cv_metrics = pd.DataFrame({"Accuracy": cv_acc, "Precision": cv_precision,
                           "Recall": cv_recall, "F1": cv_f1}, index = [0])
cv_metrics.T.plot.bar(legend = False);

plt.title("Cross VAlidation Graph")
plt.xticks(rotation = 0)
```
